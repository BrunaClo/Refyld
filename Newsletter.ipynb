```python
jtplot.style(theme='solarizedl', gridlines= '-.', context = 'notebook',figsize=(6,6))

ax1 = sns.set_style(style=None, rc=None )

fig, ax1 = plt.subplots()

sns.lineplot(data = nw_foo['Total Recipients'], marker='o', ax=ax1, )
ax2 = ax1.twinx()
plt.xlabel('Number of the week', fontsize=11.0)

plt.title('Newsletter Sent x Unsubscribed')
sns.barplot(data = nw_foo, x='week', y = 'Total Unsubs' , hue= 'week_day', alpha=0.5, ax=ax2, palette='Set1_r', dodge=False)
plt.legend(title='Day of the week', loc=2, fontsize= 10, title_fontsize=12)
plt.xlabel('Number of the week')
plt.savefig('NW Sent x unsubscribed.jpg', dpi = 200)
plt.tight_layout()

plt.show()
```
